```python
import pandas as pd
import sys
```

```python
# Our small data set
d = {'one':[1,1,1,1,1],
     'two':[2,2,2,2,2],
     'letter':['a','a','b','b','c']}

df = pd.DataFrame(d)
df
```

```python
# Create group object
one = df.groupby('letter')
one.sum()
```

```python
letterone = df.groupby(['letter','one']).sum()
letterone
```

```python
letterone.index
```

```python
letterone = df.groupby(['letter','one'], as_index=False).sum()
letterone
```
